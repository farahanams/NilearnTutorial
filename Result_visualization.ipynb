```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import seaborn as sns
import pandas as pd
```

```python
# read data
df_abc = pd.read_csv('/home/farahana/Documents/Tutorial Python/datanew/abc_cv.csv')
df_bag = pd.read_csv('/home/farahana/Documents/Tutorial Python/datanew/bag_cv.csv')
```

```python
# Concatenate both dataset
df1 = pd.concat([df_abc,df_bag])
```

```python
df1.head()
```

#### Setting the Seaborn style
rerun the code for any effect on the style

```python
sns.set()
fig=plt.figure(figsize=(12,11))
flatui = ['#f7f7f7','#cccccc','#969696','#525252']

#sns.set_palette(sns.color_palette(flatui, 8))
sns.set(style="whitegrid")
#sns.set_style("ticks")
sns.set_context("poster")
```

```python
sns.despine(trim=True)
```

```python
bp= sns.swarmplot(x="Number of components", y="Accuracy", hue="machine learning", data=df1)
plt.setp(bp.get_legend().get_title(), fontsize='18') # for legend title
plt.setp(bp.get_legend().get_texts(), fontsize='16') # for legend text
```

```python
bp= sns.swarmplot(x="Number of components", y="Accuracy", hue="Dimension reduction", data=df1)
```

```python
sns.factorplot(x="Number of components", y="Accuracy", hue="machine learning",data=df1, kind="box", size=10)
```

```python
sns.factorplot(x="Number of components", y="Accuracy", hue="Dimension reduction",data=df1, kind="box", size=10)
```

```python
sns.lmplot("Number of components", "Accuracy",data=df1, size =10)
```
