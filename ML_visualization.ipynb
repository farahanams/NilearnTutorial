```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import seaborn as sns
import pandas as pd
```

```python
#read data
df = pd.read_csv('/home/farahana/Documents/Tutorial Python/datanew/ml_cv.csv')
```

```python
df.head()
```

#### Setting the Seaborn style

```python
sns.set()
fig=plt.figure(figsize=(10,8))

sns.set(style="whitegrid")
#sns.set_style("ticks")
sns.set_context("poster")
sns.despine(trim=True)
```

### Swarmplot

#### By machine learning

```python
bp = sns.swarmplot(x="Number of components", y="Accuracy", hue="machine learning", data=df)
plt.setp(bp.get_legend().get_title(), fontsize='18') # for legend title
plt.setp(bp.get_legend().get_texts(), fontsize='16') # for legend text
```

#### By Dimension reduction

```python
bp = sns.swarmplot(x="Number of components", y="Accuracy", hue="Dimension reduction", data=df)
plt.setp(bp.get_legend().get_title(), fontsize='18') # for legend title
plt.setp(bp.get_legend().get_texts(), fontsize='16') # for legend text
```

### Pairplot

```python
sns.pairplot(df, hue="machine learning",size=5)
```

### Factorplot

#### With point figure

```python
bp= sns.factorplot(x="Number of components", y="Accuracy", col="machine learning", hue="Dimension reduction", kind="point",  data=df, size =6, aspect=0.7)
```

```python
bp= sns.factorplot(x="Number of components", y="Accuracy", hue="machine learning", col="Dimension reduction", kind="point", size=7,  data=df)
```

```python
sns.set()
fig2=plt.figure(figsize=(14,12))

sns.set(style="whitegrid")
sns.set_context("poster")
sns.despine(trim=True)
bp= sns.factorplot(x="Number of components", y="Accuracy", hue="machine learning", col="Dimension reduction", kind="bar",  data=df, size =6)
```

### Jointplot

```python
sns.jointplot(y="Number of components", x="Accuracy", kind="hex",  data=df)
```

### Lmplot

```python
sns.lmplot(x="Number of components", y="Accuracy", hue="machine learning", truncate=True,  data=df, size=8)
```
